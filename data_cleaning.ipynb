```python
import pandas as pd
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[df['Salary Estimate']!='-1']
```

```python
df.info()
```

### cleaning the salary estimate column

```python
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary.head()
```

### now to remove the K and dollar sign from the salary column

```python
minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
```

```python
minus_kd.head()
```

### create a new column for salary with hourly payment

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df[df['hourly'] == 1].head(4)
```

### create a new column for salary with employer provided payment

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df.head()
```

```python
df[df['employer_provided'] == 1].head(3)
```

```python
min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
```

### removed the 'per hour' and 'employer provided salary' and stored in min_hr

```python
min_hr
```

### remove the '-' from the min_hr

### create a min and max salary column and add it to the df table

```python
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1]))
```

```python
df.head() #check for min and max salary
```

### create a new average salary column

```python
df['average_salary'] = (df.min_salary+df.max_salary)/2
```

```python
df.head()
```

## Company name cleaning

**company name only**

```python
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis=1)
```

```python
df.head()
```

### state field

```python
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.head()
```

**jobs per state**

```python
df.job_state.value_counts()
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

### Find the age of the company

```python
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2021-x)
```

### get job description for python, r studio, excel etc

**python seeking jobs**

```python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0)
```

```python
df['python_yn'].value_counts()
```

**r studio seeking jobs**

```python
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r-studio' in x.lower() else 0)
```

```python
df.R_yn.value_counts()
```

**sparks jobs**

```python
df['spark_yn'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() or 'sparks' in x.lower() else 0)
```

```python
df.spark_yn.value_counts()
```

```python
df['excel_yn'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```

```python
df.excel_yn.value_counts()
```

```python
df['aws_yn'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df.columns
df_out = df.drop(['Unnamed: 0'],axis=1)
```

```python
df_out.head()
```

```python
df_out.to_csv('salary_data_cleaned.csv',index=False)
```
